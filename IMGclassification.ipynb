```python
import pandas as pd
import numpy as np
import os
from tensorflow.keras.applications import VGG16
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Load the training data
train_data = pd.read_csv(r"C:\Users\Dell\Desktop\train_auburn.csv")

# Path to the image directory
test_image_dir = r"C:\Users\Dell\Documents\GitHub\Auburn_Waves_CelestialChallengers\test_dataset"
train_image_dir=r"C:\Users\Dell\Documents\GitHub\Auburn_Waves_CelestialChallengers\train_dataset"
# Function to extract features using VGG16
def extract_features(img_path, model):
    img = load_img(img_path, target_size=(224, 224))
    img = img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = img / 255.0
    features = model.predict(img)
    return features.flatten()

# Load the VGG16 model
vgg16 = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Extract features for all images in the training set
features = []
labels = []
for index, row in train_data.iterrows():
    img_path = os.path.join(train_image_dir, row['File Name'])
    feature = extract_features(img_path, vgg16)
    features.append(feature)
    labels.append(row['Class'])

features = np.array(features)
labels = np.array(labels)

# Split the data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(features, labels, test_size=0.2, random_state=42)

# Train a RandomForest classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# Validate the model
y_pred = clf.predict(X_val)
print('Validation Accuracy:', accuracy_score(y_val, y_pred))

# Extract features for all images in the testing set
test_features = []
test_file_names = []
for file_name in os.listdir(test_image_dir):
    img_path = os.path.join(test_image_dir, file_name)
    if os.path.isfile(img_path) and img_path.endswith(('jpg', 'jpeg', 'png')):
        feature = extract_features(img_path, vgg16)
        test_features.append(feature)
        test_file_names.append(file_name)

test_features = np.array(test_features)

# Predict the labels for the testing set
test_predictions = clf.predict(test_features)

# Save the predictions to a CSV file
test_results = pd.DataFrame({'File Name': test_file_names, 'Class': test_predictions})
test_results.to_csv(r"C:\Users\Dell\Desktop\test_predictions.csv", index=False)
```
